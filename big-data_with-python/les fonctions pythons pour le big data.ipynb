## La fonction Lamdda en python

## La fonction map en python
En générale, on utilise une fonction map appliqué à une liste.
On utulise une fonction map quand on souhaite appliquer une fonction à une liste.
> `map(fonction_apply, liste)`

C'est une fonction intégré en python donc pas besoin de l'importer.

```python
my_liste = [1, 2, 3, 4, 5, 6, 7, 8, 9]

def squared(x):
    return x**2

squares = map(squared, my_liste)
squares
```

```python
#list(squares)
```

Pour afficher les élements contenu dans l'objet map créer : 
> On peut utiliser une `boucle for`

```python
for i in squares:
    print(i)
```

> On peut également utiliser une fonction Lambda

```python
squared1 = map(lambda x:x**2, my_liste)
squared1
```

```python
list(squared1)
```

## La fonction filter 

Il est presque pareil à la fonction map mais il permet de filtrer les éléments
> `filter(function_appy, iterable)`

```python
num = [-3, -1, -2, -1, 0, 1, 2, 3]

def positive(x):
    return x>0

num_b = filter(positive, num)
num_b
```

```python
num_pos = list(num_b)
num_pos
```

```python
object1 = filter(lambda x: x>0, num)
list(object1)
```

## La fonction reduce 

La function reduce permet de ramener à une valeur unique 
> `reduce(function(x, y), objet_iterable)`

Elle n'est pas directement intégré dans python. 
- On l'importe à partir de la librairie **functools**

```python
from functools import reduce

list_nombre = [8, 4, 5, 7, 9, 10]

"""Num plus_petit prend deux éléments, prend le plus petit 
    et le retourne en sortie"""

def plus_petit(x, y):
    if x < y:
        return x
    else:
        return y
    
reduce(plus_petit, list_nombre)
```

```python
# peut le faire directement avec la function lambda

reduce(lambda x,y: x if x<y else y, list_nombre)
```

## The Map Function

```python
import math

# Function that computes the area of a circle with a radius r
def area(r):
    """Area of a circle with radius 'r'"""
    return math.pi * (r**2)

radii = [2, 5, 7.1, 0.3, 10]
```

```python
# Methode 1: Direct Method

areas = []
for r in radii:
    a = area(r)
    #round(a, 3)
    areas.append(round(a, 3))

print(areas)
```

```python
# Methode 2 : Use "map" function

intuition_behing_map= """
The map function takes function takes two arguments
- The 1st : a function
- The 2nd : is a list, tuple or other iterable object

And the output is a map object which is actually an iterator over
the results. This is highly favorable especially when working with
large collection of data

We can then turn the map object into a liste by passing the object
to the list constructor

----------------------------------------------------------------
                         'map' function

- Data: a1, a2, a3, ..., an
- Function: f

- map(f, data)
    Returns iterator over
    f(a1), f(a2), ..., f(an)
"""

print(intuition_behing_map)
```

```python
map(area, radii)
```

```python
list(map(area, radii))
```

## The Filter Function

The filter function is use to select certain peaces of data into a liste, tuple or order collection of data. This name was used because it filtes out the data you do not need.

```python
import statistics

data = [1.3, 2.7, 0.8, 4.1, 4.3, -0.1]

avg = statistics.mean(data)
avg
```

## The Reduce Function

This function is a bit unusual.

In Python 3+, reduce is not a builtin function. It has been moved to the `functools` module 



```python
intuition_beyind_reduce = """
--------------------------------------------------------------------
                         'reduce' function intuition
--------------------------------------------------------------------

- Data: [a1, a2, a3, ..., an]
- Function: f(x, y)

- reduce(f, data)
    Step 1: val_1 = f(a1, a2)
    Step 2: val_2 = f(val_1, a3)
    Step 3: val_3 = f(val_2, a4)
    .
    .
    .
    Step n-1: val_n-1 = f(val_n-2, an)
    Return val_n-1
    
- Alternatively :
    Returns f(f(f(a1, a2), a3),a4), ..., an) 
"""
```

```python
print(intuition_beyind_reduce)
```

### Process of MapReduce for WordCount


```python
# 3. Word count program

data_words = ["AABDDDBCCABBEEEC"] 

#!usr/bin/python
"""mapper.py"""
import sys

for line in sys.stdint:
    line = line.strip()
    words = line.split()
    for word in words:
        print('%s\%s' % (word, 1))

 

#!usr/bin/python 
"""reducer.py"""
import sys 

prev_word = None
prev_count = 0

for line in sys.stdin:
    line = line.strip()
    word, count = line.split('\t')
    
    count=int(count)
    
    if prev_word == word:
        prev_count += count
    else:
        if prev_word:
            print('%s\%s' % (prev_word, prev_count))
        prev_count = count
        prev_word = word
        
if prev_word == word:
    print('%s\%s' % (prev_word, prev_count))
```
