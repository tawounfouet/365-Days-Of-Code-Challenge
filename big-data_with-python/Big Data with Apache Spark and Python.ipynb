```python
# Frist import SparkContext object for pyspark
from pyspark import SparkContext

# Create an instance of the object 
"""
A SparkContext represents the connection to a Spark cluster, and can 
be used to create RDDs, accumulators and broadcast variables on that 
cluster.
    - Since we are using the app locally we have to specify local
    - Then give our app a name
"""
sc = SparkContext("local", "first-app")
```

```python
# creating a RDD - Resilient Distributed Dataset
"""
The .parellize() methode create a distributed collection which can be
access or process on multiple nodes
"""

rdd = sc.parallelize(['C', 'C++', 'C#', 'Java', 'Python', 'Scala', 'R'])
```

```python
# Return the values of our RDD
"""
The .collect() method will return all the values of our collection
to the driver.
"""

rdd.collect()
```

```python
print(rdd.collect())
```

```python
print('Prog. Languages : %s' %(rdd.collect()))
```
