```python
BigData_lesson_4 = """
Subjet : Big Data, Hadoop with Python

Instructor : Mohammad Rafi Barakzai

Reference Books: 
    - Professional Hadoop Solutions, 
    - Understanding Big Data,
    - "HADOOP: The definitive Guide" and Hadoop and Python

---------------------------------------------------------------------
                            Lesson 4
                            
Topics to be covered:
1. Hadoop architechture
2. Hadoop services 
3. Python installation
4. Example

----------------------------
1. Hadoop architecture 
        Hadoop has the following two main components:

i) HDFS
        Hadoop Distributed File System (HDFS): Stores large volume of
        data over a network of clusters(computers).
        It's the same as a Relatonal Database Management System (DBMS)
        
        HDFS is designed for fast, fault-tolerant processing, thus 
        enabling the use of inexpensive storage hardware.
        
        MapReduce Programming Model is used to process and deal with 
        large data volumes.

We set in the previous lecture that in normal distributed system,
the data will be move to the server and the execution will be done 
their. This process is expensive because we have to increase the number
of processors if needed, increase the size of, Hard Disk Drive, RAM.
    But in HDFS the data will be stored on local computers or machines; 
only the control will be same to that of a normal system but the
Computation and the storage is done on the same local node -> That is 
why it's call unexpensive storage hardware.

Unique features of HDFS
- Handling large datasets
- Fault-tolerance
- Streaming access to data
- Simple data consistency
    
    
ii) YARN
        Yet Another Resource Negociator (YARN) : It is Hadoop processing
        layer, operating system of Hadoop, resource manager or scheduler
        
        YARN is designed to manage all resources in a Hadoop cluster.
        
        YARN is a framework for managing distributed applications 
        executed on multiple machines within a network.
        
        It contains resource manager and job scheduler for both 
        resources and jobs to handle.
            The job scheduler is there to handle applications 
        and asign the jobs(or tasks) to the resources.
figure 
        

In Hadoop clusters, there are two types of nodes we have. Some of the
nodes are call the master nodes and the remaining nodes are call the
workers nodes.
- Masters node are nothing but the main server 
- Workers nodes are all the remaining server and this nodes are control
  by the master node
So some of the services will be running on the Master Nodes and others 
will be running on the workers nodes
    
---------------------
2. Hadoop Services :
        There is a set of HDFS related services that take care of the 
        storage related work, and there is a set of computational 
        services related to YARN.
        
i) HDFS services
        - Namenode : The service runs on master node and maintains 
                     metadata pertaining to HDFS storage :
                        - File system
                        - File's location
                     When a client seeks to read or write to HDFS,
                     it contacts the Namenode serice, which provides the
                     client information about the location of the file
                     in HDFS.
            
        - Secondary Namenode or Standby Namenode : 
                    The service runs on master node and relieves the 
                    burden of the critical Namenode by performing tasks
                    such as updating the metadata file
        
        - Datanodes : The service runs on worker nodes that stores the 
                    HDFS data Blocks on the linux file system. The 
                    Datanodes keep in contact with the Namenode and 
                    update the latter with all the changes that occur 
                    in the file system.

ii) YARN Services
        - Resource Manager : Runs on master node, allocating resouces and
                   scheduling jobs on the worker nodes.
                   
        - Application Manager : Runs on master node, coordinating the 
                  execution of application in the cluster.
                  
        - Node Manader : Runs on worker nodes and reports to resource
                    manager the health and task running.
    
                    
                    
3. Python Installation

        Python is a popular and simple programming language.
        Features of Python programming language
        - Simple Syntax
        - Easy to learn
        - Cross Plateform
        - Open Source and Free
        - Large Standard Libraries
        - Advance Features

How to install ?

Linux :
- Sudo apt-get update
- sudo apt-get install python
- python --version
- python -V
        
        
        
"""
```
